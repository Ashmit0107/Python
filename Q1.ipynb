```python
##find the smallest number in an array

def find_smallest(arr):
    smallest=arr[0]
    for i in range(len(arr)):
        if smallest>arr[i]:
            smallest=arr[i]
    return smallest
##main
num=int(input("Enter the number of elements:"))
arr=[]
for i in range(num):
    x=int(input(f"Enter {i+1}th element:"))
    arr.append(x)
print("Smallest element:",find_smallest(arr))
```
