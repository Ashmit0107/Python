```python
# smallest number in the array
def smallest_array(arr):
    smallest=arr[0]
    for i in range(1,len(arr)):
        if arr[i]<smallest:
            smallest=arr[i]
    return smallest
print("Smallest number in the array [98,6,33,7,10,1]:",smallest_array([98,6,33,7,10,1]))
```

```python
# second smallest number in the array
def second_smallest_array(arr):
    smallest=arr[0]
    second_smallest=arr[1]
    for i in range(1,len(arr)):
        if arr[i]<smallest:
            smallest=arr[i]
        if smallest>second_smallest:
            second_smallest=smallest
    return second_smallest
print("Smallest number in the array [98,6,33,7,10,1]:",second_smallest_array([98,6,33,7,10,1]))
```

```python
# reverse a given array
def reverse_array(arr):
    start=0
    end=len(arr)-1
    while start<end:
        arr[start],arr[end]=arr[end],arr[start]
        start+=1
        end-=1
    return arr
print("Reverse of the array [98,6,33,7,10,1]:",reverse_array([98,6,33,7,10,1]))
```

```python
# count the frequency of each element
def find_freq(arr):
    freq={}
    for num in arr:
        if num in freq:
            freq[num]+=1
        else:
            freq[num]=1
    return freq
print("Frequency of the numbers in the array [98,7,6,33,6,7,1,10,1,7]:",find_freq([98,7,6,33,6,7,1,10,1,7]))
```

```python
# arrange the given array in increasing-decreasing(zig-zag order)
def zigzag(arr):
    arr.sort()
    for i in range(0,len(arr)-1,2):
        arr[i],arr[i+1]=arr[i+1],arr[i]
    return arr
print("Waveform of [5,6,3,4,1,2]:",zigzag([5,6,3,4,1,2]))
```

```python
# remove duplicates
def remove_duplicates(arr):
    seen=[]
    for i in range(len(arr)):
        if arr[i] not in seen:
            seen.append(arr[i])
    return seen
print("Remove duplicates of [3,7,3,1,1,2]:",remove_duplicates([3,7,3,1,1,2]))
print("Remove duplicates of [0,0,1,2,3,7,7]:",remove_duplicates([0,0,1,2,3,7,7]))
```

```python
# rotate array by k
def rotate_array(arr,k):
    if k==0:
        return arr
    k=k%len(arr)
    left=arr[:k]
    right=arr[k:]
    arr[:]=right+left
    return arr
print("[1,2,3,4,5,6,7] is rotated by 33:",rotate_array([1,2,3,4,5,6,7],33))
```

```python
# in how many transformations shall arr1 will become arr2
def transform(arr1, arr2):
    for i in range(len(arr1)):
        if arr1[i:] + arr1[:i] == arr2:
            return i
    return -1

arr1 = [1, 2, 3, 4, 5]
arr2 = [3, 4, 5, 1, 2]

print(transform(arr1, arr2))
```

```python
# rank replacement
def rank_replace(arr):
    s = sorted(arr)
    result = []
    for i in arr:
        result.append(s.index(i) + 1)
    return result
arr = [40, 10, 20, 30]
print(rank_replace(arr))
```

```python
def sort_frequency(arr):
    freq = {}
    for x in arr:
        freq[x] = freq.get(x, 0) + 1
    return sorted(arr, key=lambda x: (-freq[x], x))

arr = [2,5,2,8,5,2,2,6,8,8,8]
print(sort_frequency(arr))
```

```python
# palindrome check
def palindrome(s):
    s=str(s)
    return s==s[::-1]
print("is madam palindrome? ",palindrome("madam"))
print("is 12321 palindrome? ",palindrome(12321))
print("is 19 palindrome? ",palindrome(19))
```

```python
def repeated_word(s):
    words = s.split()
    ans = ""
    mx = 0
    for word in words:
        repeat = 0
        for ch in word:
            repeat = max(repeat, word.count(ch))
        if repeat > mx:
            mx = repeat
            ans = word
    return ans

print(repeated_word("I love programming committee apple"))
```

```python
# Question:
# Search an element in an array using Linear Search.

def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i+1
    return -1

n = int(input())
arr = list(map(int, input().split()))
key = int(input())
print(arr)
print(linear_search(arr, key))

# Input
# 5
# 10 20 30 40 50
# 30

# Output
# 2
```

```python
# Question:
# Search an element in a sorted array using Binary Search.

def binary_search(arr, key):
    low, high = 0, len(arr)-1
    while low <= high:
        mid = (low+high)//2
        if arr[mid] == key:
            return mid+1
        elif arr[mid] < key:
            low = mid+1
        else:
            high = mid-1
    return -1

n = int(input())
arr = list(map(int, input().split()))
key = int(input())
print(binary_search(arr, key))

# Input
# 5
# 10 20 30 40 50
# 40

# Output
# 4
```

```python
# Question:
# Sort an array using Bubble Sort.

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

n = int(input())
arr = list(map(int, input().split()))
print(*bubble_sort(arr))

# Input
# 5
# 5 2 4 1 3

# Output
# 1 2 3 4 5
```
