```python
##find the smallest number in an array
def find_smallest(arr):
    smallest=arr[0]
    for i in range(len(arr)):
        if smallest>arr[i]:
            smallest=arr[i]
    return smallest
##main
num=int(input("Enter the number of elements:"))
arr=[]
for i in range(num):
    x=int(input(f"Enter {i+1}th element:"))
    arr.append(x)
print("Array is:",arr)
print("Smallest element:",find_smallest(arr))
```

```python
##second smallest & second largest element in an array
def second_smallest_largest(arr):
    firstsmall=secondsmall=float('inf')
    firstlarge=secondlarge=float('-inf')
    for num in arr:
        #smallest
        if num<firstsmall:
            secondsmall=firstsmall
            firstsmall=num
        elif firstsmall < num < secondsmall:
            secondsmall=num
        #largest
        if num>firstlarge:
            secondlarge=firstlarge
            firstlarge=num
        elif firstlarge > num > secondlarge:
            secondlarge=num
    return secondlarge,secondsmall
print("2nd largest & 2nd smallest are",second_smallest_largest([5,2,9,1,7]))
```

```python
##reverse a given array
def reverse_array(arr):
    start=0
    end=len(arr)-1
    while start<end:
        arr[start],arr[end]=arr[end],arr[start]
        start+=1
        end-=1
    return arr
print("Reverse of the array is",reverse_array([5,2,9,1,7]))
```

```python
##count frequency of each element in an array
def count_freq(arr):
    freq={}
    for num in arr:
        if num in freq:
            freq[num]+=1
        else:
            freq[num]=1
    return freq
print("Frequency of elements in the array is",count_freq([1,2,2,3,3,3,4,4,4,4,5]))
```
