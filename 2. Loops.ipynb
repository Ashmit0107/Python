```python
## Sum of n natural numbers using while loop
num=float(input("Enter number of natural numbers to be considered:"))
sum=0
count=1 ## starting number
while count<=num:
    sum=sum+count
    count=count+1
print("Sum of",num,"natural numbers are:",sum)

```

```python
## Sum of n natural numbers using for loop
num=int(input("Enter number of natural numbers to be considered:"))
sum=0
count=1 ## starting number
for i in range(num+1):
    sum=sum+i
print("Sum of",num,"natural numbers are:",sum)
```

```python
## Check for prime numbers from 1-100
for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
