```python
# find all symmetric pairs in array like (a,b) and (b,a)
def symmetric_pairs(arr):
    d={} # dictionary to check the frequency of pairs
    result=[] # list storing the symmetric pairs
    for a,b in arr:
        if (b,a) in d:
            result.append((a,b))
        else:
            d[(a,b)]=1
    return result
# main
arr=[(1,2),(3,4),(5,6),(2,1),(6,5)]
print("The given array is ",arr)
print("The given symmetric pairs in the array is ",symmetric_pairs(arr))
```

```python
# replace each element of an array by its rank
def rank_array(arr):
    temp = sorted(set(arr))   # sorted unique elements
    rank = {}
    for i in range(len(temp)):
        rank[temp[i]] = i + 1
    result = []
    for i in arr:             # original order
        result.append(rank[i])
    return result

# main
arr = [40,5,8,67,77,8,5]
print("The given array is", arr)
print("Rank array is", rank_array(arr))
```
