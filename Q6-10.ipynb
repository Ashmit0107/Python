```python
##rearrange array in increasing decreasing order(zig zag/wave)
def rearrange_wave(arr):
    arr.sort()
    for i in range(1,len(arr)-1,2):
        arr[i],arr[i+1]=arr[i+1],arr[i]
    return arr
print("Original arrray is 4,3,7,8,6,2")
print("Waveform is",rearrange_wave([4,3,7,8,6,2]))
```

```python
## Rotate elements by K elements - Block Swamp algorithm
def block_swamp_rotate(arr,k):
    n=len(arr)
    k=k%n
    if k==0: return arr
    left=arr[:k]
    right=arr[k:]
    arr[:]=right+left
    return arr
print("Rotated by K-block",block_swamp_rotate([1,2,3,4,5,6,7],2))
```

```python
#median of given array
def find_median(arr):
    arr.sort()
    n=len(arr)
    mid=n//2
    if n%2==0:
        return (arr[mid-1]+arr[mid])/2
    else:
        return arr[mid]

# main
print("Median of the array [5,2,9,1,7] is ",find_median([5,2,9,1,7]))
```
