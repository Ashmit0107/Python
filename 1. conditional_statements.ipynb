```python
## Assignment
## Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation=input("Enter operation(+,-,*,/):")

##Perform the operation as user-input
if operation == "+":
    result= num1 + num2
elif operation == "-":
    if num1>num2:
        result = num1-num2
    else:
        result = num2-num1
elif operation == "*":
    result= num1 * num2
elif operation == "/":
    if num2!=0:
        result = num1/num2
    else:
        result = "ERROR: DIVISON by ZERO"
else:
    result = "INVALID OPERATION"

print("Result is ",result)
```
