```python
# check if a string is palindrome or not

def is_string_palindrome(s):
    return s==s[::-1]
print("Check the word 'MADAM' for palindrome:",is_string_palindrome("madam"))
```

```python
# count the number of spaces, vowels and consonants
def consonants_spaces_vowels(s):
    vowels="AEIOUaeiou"
    vowel_count=0
    consonant_count=0
    space_count=0
    for ch in s:
        if ch==' ':
            space_count+=1
        elif ch.isalpha():
            if ch in vowels:
                vowel_count+=1
            else:
                consonant_count+=1
    return space_count,vowel_count,consonant_count

print("Spaces, vowels, consonants of the given word, 'Hello World' is:",consonants_spaces_vowels("Hello World"))
```

```python
# return the ASCII value of a character
def ascii_value(ch):
    return ord(ch)
print("ASCII of A is:",ascii_value("A"))
```

```python
# remove all the vowels
def remove_vowels(s):
    vowels="AEIOUaeiou"
    result=""
    for ch in s:
        if ch not in vowels:
            result+=ch
    return result

print("without vowels, 'Hello World' is:",remove_vowels("Hello World"))
```

```python
# remove all the spaces
def remove_spaces(s):
    result=""
    for ch in s:
        if ch!=' ':
            result+=ch
    return result

print("without space, 'Hello World' is:",remove_spaces("Hello World"))
```

```python
# remove all the non-alphabets
def remove_nonalphabets(s):
    result=""
    for ch in s:
        if ch.isalpha():
            result+=ch
    return result

print("Removing all the non-alphabets, '7Hell15#o World89' is:",remove_nonalphabets("7Hell15#o World89"))
```

```python
# sum of the numbers in a string
def sum_num(s):
    l=[]
    sum=0
    for char in s:
        if char.isdigit():
            l.append(int(char))
    for i in range(len(l)):
        sum+=l[i]
    return sum
print("Sum of numbers in a string:",sum_num("abc12de34fg5"))
```

```python
# capitalize first and last letter/character of a string
def cap(s):
    if len(s) <= 1:
        return s.upper()
    return s[0].upper() + s[1:-1] + s[-1].upper()

print(cap("python"))
```

```python
# calculate first non-repeating character in a string
def first_non_repeat(s):
    for ch in s:
        if s.count(ch) == 1:
            return ch
    return -1

print(first_non_repeat("aabbcdde"))
```

```python
# check if 2 strings are anagram
# Anagram: Two strings are anagrams if they contain 
# the same letters with the same frequency, 
# but possibly in a different order

def anagram(s1,s2):
    s1=s1.replace(" ","").lower()
    s2=s2.replace(" ","").lower()
    return sorted(s1)==sorted(s2)
print("Is anagram (listen & silent):",anagram("listen","silent"))
```
