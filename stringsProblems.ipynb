```python
# check if a string is palindrome or not

def is_string_palindrome(s):
    return s==s[::-1]
print("Check the word 'MADAM' for palindrome:",is_string_palindrome("madam"))
```

```python
# count the number of spaces, vowels and consonants
def consonants_spaces_vowels(s):
    vowels="AEIOUaeiou"
    vowel_count=0
    consonant_count=0
    space_count=0
    for ch in s:
        if ch==' ':
            space_count+=1
        elif ch.isalpha():
            if ch in vowels:
                vowel_count+=1
            else:
                consonant_count+=1
    return space_count,vowel_count,consonant_count

print("Spaces, vowels, consonants of the given word, 'Hello World' is:",consonants_spaces_vowels("Hello World"))
```

```python
# return the ASCII value of a character
def ascii_value(ch):
    return ord(ch)
print("ASCII of A is:",ascii_value("A"))
```
