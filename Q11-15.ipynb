```python
#remove duplicates from a array & sort it
def remove_duplicates(arr):
    n=len(arr)
    result=[arr[0]]
    for i in range(1,n):
        if arr[i] not in result:
            result.append(arr[i])
    result.sort()
    return result

print("Duplicates removed from array, [1,2,8,1,1,2,6,9,4,1], & sorted as well is ", remove_duplicates([1,1,2,8,1,1,2,6,9,4,1]))
# this code will be same for unsorted as well
```

```python
# adding an element at a given position
def adding_element(arr,position,value):
    arr.insert(position,value)
    return arr
print("After adding element to [5,2,9,1,7] is",adding_element([5,2,9,1,7],3,4))
```

```python
# find all repeating elements in an array
def frequency_elements(arr):
    freq={}
    for num in arr:
        if num in freq:
            freq[num]+=1
        else:
            freq[num]=1
    return freq
def repeating_elements(arr):
    freq=frequency_elements(arr)
    repeated=[]
    for num in freq:
        if freq[num]>1:
            repeated.append(num)
    return repeated
def unique_elements(arr):
    freq=frequency_elements(arr)
    unique=[]
    for num in freq:
        if freq[num]==1:
            unique.append(num)
    return unique
print("Frequency of the elements of the array, [1,1,2,8,1,1,2,6,9,4,1], are ", frequency_elements([1,1,2,8,1,1,2,6,9,4,1]))
print("Repeating elements of the array, [1,1,2,8,1,1,2,6,9,4,1], are ", repeating_elements([1,1,2,8,1,1,2,6,9,4,1]))
print("Unique elements of the array, [1,1,2,8,1,1,2,6,9,4,1], are ", unique_elements([1,1,2,8,1,1,2,6,9,4,1]))
```
