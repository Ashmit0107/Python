```python
# reverse a number
def reverse_number(num):
    rev=0
    while num>0:
        digit=num%10
        rev=rev*10+digit
        num=num//10
    return rev

n=int(input("Enter entire number:"))
print(f"Reverse of",n,"is",reverse_number(n))
```

```python
# check for prime number
# a prime number is divisible only by 1 & itself 

def prime(num):
    if num<2:
        return False
    for i in range(2,int(num**0.5)+1):
        if num%i==0:
            return False
    return True

n=int(input("Enter number:"))
print(f"Check for Prime of",n,"is",prime(n))
```

```python
# print all the factors of a number and also count the total number of factors
def factors(n):
    count = 0
    print("Factors are:")
    for i in range(1, int(n**0.5) + 1):
        if n % i == 0:
            print(i, end=" ")
            count += 1
            if i != n // i:
                print(n // i, end=" ")
                count += 1
    print("\nTotal Factors =", count)


# Main
n = int(input())
factors(n)
```

```python
# ARMSTRONG NUMBER
# armstrong no. is the number which is equal to the sum of the cube of it's digits
def armstrong(num):
    original=num
    arr=[]
    cube_sum=0
    while num>0:
        digit=num%10
        num=num//10
        arr.append(digit)
    for i in range(len(arr)):
        cube_sum=cube_sum+(arr[i]**3)
    if cube_sum==original:
        return True
    else:
        return False

# Main
n = int(input())
print(f"Number is",n,"& it's check for Armstrong is",armstrong(n))
```

```python
# PALINDROME NUMBER
# palindrome number is the number which is equal to it's reverse 
def reverse_number(num):
    rev=0
    while num>0:
        digit=num%10
        rev=rev*10+digit
        num=num//10
    return rev
def palindrome(num):
    reverse=reverse_number(num)
    if reverse==num:
        return True
    else:
        return False

# Main
n = int(input())
print(f"Number is",n,"& it's check for Palindrome is",palindrome(n))
```

```python
# check for perfect number (sum of proper factors (excluding the number itself) is equal to the number)
def perfect(num):
    sum = 0
    for i in range(1, num):
        if num % i == 0:
            sum += i
    return sum == num

# Main
n = int(input())
if perfect(n):
    print("Perfect Number",n)
else:
    print("Not a Perfect Number",n)
```

```python
def leapYear(year):
    if year % 400 == 0:
        return True
    elif year % 100 == 0:
        return False
    elif year % 4 == 0:
        return True
    else:
        return False


# Main
year = int(input())
if leapYear(year):
    print(year,"is a leap Year")
else:
    print(year,"is not a Leap Year")
```

```python
# maximum and minimum numbers digit in a number
def max_min(num):
    digit=str(num)
    max_digit=digit[0]
    min_digit=digit[0]
    for d in digit:
        if d>max_digit:
            max_digit=d
        if d<min_digit:
            min_digit=d
    return int(max_digit),int(min_digit)

print("Max,min digit of 39284",max_min(39284))
```

```python
# print fibonacci series upto n
def fibonacci(n):
    series=[]
    a,b=0,1
    for i in range(n):
        series.append(a)
        a,b=b,a+b
    return series
print("Fibonacci first 10 terms",fibonacci(10))
```

```python
# factorial of a number
def factorial(n):
    result=1
    for i in range(2,n+1):
        result*=i
    return result
print("Factorial of 7 is",factorial(7))
```

```python
# factors of a given number
def find_factors(n):
    factors=[]
    for i in range(1,n+1):
        if n%i==0:
            factors.append(i)
    return factors
print("Factors of 36 is",find_factors(36))
```

```python
# prime factors of a given number
def prime_factor(n):
    factors=[]
    divisor=2
    while n>1:
        while n % divisor ==0:
            factors.append(divisor)
            n=n//divisor
        divisor+=1
    return factors
print("Prime factors of 36 is",prime_factor(36))
```

```python
# check if a number is a strong number
# (sum of the factorials of the digits of the numbe == the number, eg 145)

def strong_no(num):
    original=num
    sum=0
    while num>0:
        digit=num%10 #extracting the last digit
        fact=1 #finding the factorial of the last digit
        for i in range(1,digit+1):
            fact*=i
        sum+=fact #adding the factorial to the sum
        num=num//10 #reducing the number by removing the last digit
    return sum==original

# Main
n = int(input())
if strong_no(n):
    print(n,"is a strong number")
else:
    print(n,"is not a strong number")
```

```python
# check if a number is automorphic 
# square of a number ends with the same number eg, 25 & 625

def is_Automorphic(n):
    square=n*n
    return str(square).endswith(str(n))
print("Is 25 automorphic?",is_Automorphic(25))
print("Is 11 automorphic?",is_Automorphic(11))
```

```python
# GCD & LCM of 2 numbers
def GCD(a,b):
    while b:
        a,b=b,a%b
    return a
def LCM(a,b):
    return (a*b)//GCD(a,b)
print("GCD of 36 & 60:",GCD(36,60))
print("LCM of 36 & 60:",LCM(36,60))
```

```python
# check for Harshad Number 
# (a number divisible by sum of it's digits)

def check_Harshad(num):
    digit_sum=0
    temp=num
    while num>0:
        digit_sum+=num%10
        num=num//10
    return temp==digit_sum

n=int(input("Enter number to be checked:"))
print(f"Is",n,"a harshad number?",check_Harshad(n))
```
