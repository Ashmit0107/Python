```python
# show even & odd numbers from an array
def even_odd(arr):
    even=[]
    odd=[]
    for i in range(len(arr)):
        if arr[i]%2==0:
            even.append(arr[i])
            print("Even:",arr[i])
        else:
            odd.append(arr[i])
            print("Odd:",arr[i])
arr = [40,5,8,67,77,8,5]
print("The given array is", arr)
even_odd(arr)
```

```python
# count number of digits 
def count_digits(num):
    count=0
    while num>0:
        num=num//10
        count+=1
    return count

n=int(input("Enter entire number:"))
print(f"No. of digits in",n,"is",count_digits(n))
```

```python
# reverse a number
def reverse_number(num):
    rev=0
    while num>0:
        digit=num%10
        rev=rev*10+digit
        num=num//10
    return rev

n=int(input("Enter entire number:"))
print(f"Reverse of",n,"is",reverse_number(n))
```
