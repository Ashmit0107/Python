```python
# show even & odd numbers from an array
def even_odd(arr):
    even=[]
    odd=[]
    for i in range(len(arr)):
        if arr[i]%2==0:
            even.append(arr[i])
            print("Even:",arr[i])
        else:
            odd.append(arr[i])
            print("Odd:",arr[i])
arr = [40,5,8,67,77,8,5]
print("The given array is", arr)
even_odd(arr)
```
