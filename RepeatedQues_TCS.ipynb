```python
# Replace elements by it's rank in the array
def replaceWithRank(arr):
    rank_map={}
    rank=1
    for num in sorted(arr):
        if num not in rank_map:
            rank_map[num]=rank
            rank+=1
    return [rank_map[num] for num in arr]

# main
n=int(input("Enter number of elements in the array:"))
arr=list(map(int,input().split()))
print("Given array is ", arr)
print("Rank of the array is", replaceWithRank(arr))
```

```python
# Remove characters from first string present in second string
def removeChars(str1,str2):
    removeSet=set(str2)
    remove=[]
    for chars in str1:
        if chars not in removeSet:
            remove.append(chars)
    return "".join(remove)

# main
str1=input("Enter string 1:").strip()
str2=input("Enter string 2:").strip()
print("Characters from string1 & string2 are",str1,str2)
print("After removing characters from string1 belonging to string2")
print(removeChars(str1,str2))
```

```python
# find all non-repeating elements in an array
def nonrepeatElements(arr):
    freq={} ; result=[]
    for num in arr:
        if num not in freq:
            freq[num]=1
        else:
            freq[num]+=1
    #result=[]
    for num in arr:
        if freq[num]==1:
            result.append(num)
    return result
# main
n=int(input("Enter number of elements in the array:"))
arr=list(map(int,input().split()))
print("Given array is ", arr)
print("All the non-repeating elements in the array is", nonrepeatElements(arr))
```

```python
# if a message contains 3 consecutive identical characters it is flagged as spam else safe
def safeORspam(str1):
    for i in range(len(str1) - 2):
        if str1[i] == str1[i+1] == str1[i+2]:
            return "SPAM"
    return "SAFE"

# Main
str1 = input("Enter string/message: ").strip()
print("Characters from string are", str1)
print("Message is", safeORspam(str1))
```

```python
# library fine collection 
# input: N-space separated integers 
# and if any of it is more than K, 
# then 1 unit fine per day
def library_fine(arr,k):
    fine=0
    for days in arr:
        if days>k:
            fine+=days-k
    return fine
#main
n=int(input("Enter no of books borrowed:"))
arr=list(map(int,input().split()))
k=int(input("Enter no of days it can be kept for free:"))
print("Given array is ", arr)
print("Total fine:", library_fine(arr,k))
```

```python
# TCS NQT STYLE
# bus passenger tracking, 
# there are total N stoppages 
# at each stoppage some people get-off & some board the bus
# find the maximum people in the bus at a time

class Solution: 
    def maxPassenger(self,n,arr):
        curr=0
        maxi=0
        for off,on in arr:
            curr=curr-off+on
            if curr>maxi:
                maxi=curr
        return maxi

#main code
n=int(input("Enter total number of stoppages:"))
arr=[]
for i in range(n):
    off,on=list(map(int,input().split()))
    arr.append([off,on])
obj=Solution()
print(obj.maxPassenger(n,arr))

# 4 stoppages & off-on passengers list is
# 0 3
# 2 5
# 4 2
# 4 0
```

```python
#Question 1: Coin Pair Indices 
# You are given an array of integers 
# representing coin values and an integer target value. 
# Find two different indices such that the sum of the coin values at those indices 
# equals the target value.
# Return the indices of the two coins.
# Example:
# Input: coins = [2, 7, 11, 15] & target = 9 
# Output:[0, 1]
# Explanation:coins[0] + coins[1] = 2 + 7 = 9

class Solution:
    def coinPairindices(self,coins,target):
        n=len(coins)
        for i in range(n):
            for j in range(i+1,n):
                if coins[i]+coins[j]==target:
                    return[i,j]

coins = [2, 7, 11, 15]
target = 9 
obj=Solution()
print(obj.coinPairindices(coins,target))
```

```python
# Question 2: Merge Two Arrays into a Sorted Array
# You are given two arrays:
# Array A is sorted in ascending order. 
# Array B is sorted in descending order. 
# Create a new array containing all elements 
# from both arrays such that the final array 
# is sorted in ascending order.
# Example
# Input: A = [1, 3, 5, 7]
# B = [10, 8, 6, 4]
# Output: [1, 3, 4, 5, 6, 7, 8, 10]


class Solution:
    def mergeArray(self,A,B):
        C = A + B
        C.sort()
        return C

A = [1, 3, 5, 7]
B = [10, 8, 6, 4]
obj=Solution()
print(obj.mergeArray(A,B))
```

```python
# Question:
# You are given an array of integers and a target value.
# Find the index of the first occurrence of the target value in the array.
# If the target is not present in the array, return -1.

class Solution:
    def findIndex(self, arr, target):
        for i in range(len(arr)):
            if arr[i] == target:
                return i
        return -1


arr = [10, 20, 30, 40, 30, 50]
target = 30

obj = Solution()
print(obj.findIndex(arr, target))
```
