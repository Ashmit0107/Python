```python
# Replace elements by it's rank in the array
def replaceWithRank(arr):
    rank_map={}
    rank=1
    for num in sorted(arr):
        if num not in rank_map:
            rank_map[num]=rank
            rank+=1
    return [rank_map[num] for num in arr]

# main
n=int(input("Enter number of elements in the array:"))
arr=list(map(int,input().split()))
print("Given array is ", arr)
print("Rank of the array is", replaceWithRank(arr))
```

```python
# Remove characters from first string present in second string
def removeChars(str1,str2):
    removeSet=set(str2)
    remove=[]
    for chars in str1:
        if chars not in removeSet:
            remove.append(chars)
    return "".join(remove)

# main
str1=input("Enter string 1:").strip()
str2=input("Enter string 2:").strip()
print("Characters from string1 & string2 are",str1,str2)
print("After removing characters from string1 belonging to string2")
print(removeChars(str1,str2))
```

```python
# find all non-repeating elements in an array
def nonrepeatElements(arr):
    freq={} ; result=[]
    for num in arr:
        if num not in freq:
            freq[num]=1
        else:
            freq[num]+=1
    #result=[]
    for num in arr:
        if freq[num]==1:
            result.append(num)
    return result
# main
n=int(input("Enter number of elements in the array:"))
arr=list(map(int,input().split()))
print("Given array is ", arr)
print("All the non-repeating elements in the array is", nonrepeatElements(arr))
```

```python
# if a message contains 3 consecutive identical characters it is flagged as spam else safe
def safeORspam(str1):
    for i in range(len(str1) - 2):
        if str1[i] == str1[i+1] == str1[i+2]:
            return "SPAM"
    return "SAFE"

# Main
str1 = input("Enter string/message: ").strip()
print("Characters from string are", str1)
print("Message is", safeORspam(str1))
```

```python
# library fine collection 
# input: N-space separated integers 
# and if any of it is more than K, 
# then 1 unit fine per day
def library_fine(arr,k):
    fine=0
    for days in arr:
        if days>k:
            fine+=days-k
    return fine
#main
n=int(input("Enter no of books borrowed:"))
arr=list(map(int,input().split()))
k=int(input("Enter no of days it can be kept for free:"))
print("Given array is ", arr)
print("Total fine:", library_fine(arr,k))
```

```python
# TCS NQT STYLE
# bus passenger tracking, 
# there are total N stoppages 
# at each stoppage some people get-off & some board the bus
# find the maximum people in the bus at a time

class Solution: 
    def maxPassenger(self,n,arr):
        curr=0
        maxi=0
        for off,on in arr:
            curr=curr-off+on
            if curr>maxi:
                maxi=curr
        return maxi

#main code
n=int(input("Enter total number of stoppages:"))
arr=[]
for i in range(n):
    off,on=list(map(int,input().split()))
    arr.append([off,on])
obj=Solution()
print(obj.maxPassenger(n,arr))

# 4 stoppages & off-on passengers list is
# 0 3
# 2 5
# 4 2
# 4 0
```
