```python
# Return squares of a sorted array, still sorted 
def sorted_square(arr):
    arr2=[]
    for i in range(len(arr)):
        arr2.append(arr[i] ** 2)
    arr2.sort()
    return arr2

print(sorted_square([-4, -1, 0, 3, 10]))  # Output: [0, 1, 9, 16, 100]
```

```python
# Find all unique triplets in the array that sum to zero.
def three_sum(arr):
    arr2 = []
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            for k in range(i + 2, len(arr)):
                if arr[i] + arr[j] + arr[k] == 0:   # Check if the sum is 0
                    triplet = sorted([arr[i], arr[j], arr[k]])
                    if triplet not in arr2:   # Add only if not already present
                        arr2.append(triplet)
    return arr2

print(three_sum([-1, 0, 1, 2, -1, -4]))
```

```python
# Find the triplet sum closest to a given target.
def three_sum(arr, target):
    closest = arr[0] + arr[1] + arr[2]
    ans = [arr[0], arr[1], arr[2]]
    for i in range(len(arr)):
        for j in range(i + 1, len(arr)):
            for k in range(j + 1, len(arr)):
                s = arr[i] + arr[j] + arr[k]
                if abs(s - target) < abs(closest - target):
                    closest = s
                    ans = [arr[i], arr[j], arr[k]]
    return ans

print(three_sum([-1, 0, 1, 2, -1, -4], -3))
```

```python
# Count triplets (i<j<k) whose sum is smaller than a given value

def count_triplets(arr,target):
    count=0
    for i in range(len(arr)):
        for j in range(i+1,len(arr)):
            for k in range(j+1,len(arr)):
                if arr[i] + arr[j] + arr[k] < target:
                    count += 1
    return count

print(count_triplets([-2, 0, 1, 3], 2))   # Output: 2
```

```python
# Find all unique quadruplets whose sum equals target
def four_sum(arr, target):
    ans = []
    n = len(arr)
    for i in range(n):
        for j in range(i+1, n):
            for k in range(j+1, n):
                for l in range(k+1, n):
                    if arr[i]+arr[j]+arr[k]+arr[l] == target:
                        q = sorted([arr[i],arr[j],arr[k],arr[l]])
                        if q not in ans:
                            ans.append(q)
    return ans
print(four_sum([1,0,-1,0,-2,2], 0))
```

```python
# Compare two strings after treating # as backspace
def compare(s, t):
    def build(x):
        st = []
        for ch in x:
            if ch == "#":
                if st:
                    st.pop()
            else:
                st.append(ch)
        return "".join(st)
    return build(s) == build(t)

print(compare("ab#c", "ad#c"))   # True
```
