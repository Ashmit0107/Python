```python
##find the smallest number in an array

def find_smallest(arr):
    smallest=arr[0]
    for i in range(len(arr)):
        if smallest>arr[i]:
            smallest=arr[i]
    return smallest
##main
num=int(input("Enter the number of elements:"))
arr=[]
for i in range(num):
    x=int(input(f"Enter {i+1}th element:"))
    arr.append(x)
print("The entered Array is:",arr)
print("Smallest element:",find_smallest(arr))
```

```python
## Assignment
## Simple Calculator

num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
operation=input("Enter operation(+,-,*,/):")

##Perform the operation as user-input
if operation == "+":
    result= num1 + num2
elif operation == "-":
    if num1>num2:
        result = num1-num2
    else:
        result = num2-num1
elif operation == "*":
    result= num1 * num2
elif operation == "/":
    if num2!=0:
        result = num1/num2
    else:
        result = "ERROR: DIVISON by ZERO"
else:
    result = "INVALID OPERATION"

print("Result is ",result)
```

```python
## Sum of n natural numbers using while loop
num=float(input("Enter number of natural numbers to be considered:"))
sum=0
count=1 ## starting number
while count<=num:
    sum=sum+count
    count=count+1
print("Sum of",num,"natural numbers are:",sum)
```

```python
## Sum of n natural numbers using for loop
num=int(input("Enter number of natural numbers to be considered:"))
sum=0
count=1 ## starting number
for i in range(num+1):
    sum=sum+i
print("Sum of",num,"natural numbers are:",sum)
```

```python
## Check for prime numbers from 1-100
for num in range(1,101):
    if num>1:
        for i in range(2,num):
            if num%i==0:
                break
        else:
            print(num)
```
